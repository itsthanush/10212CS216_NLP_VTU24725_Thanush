```python
import nltk
import spacy

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger_eng')

nlp=spacy.load("en_core_web_sm")

text="i am thanush and i am studing at veltech clg"

text_lower=text.lower()

tokens=word_tokenize(text_lower)

print("Tokens:",tokens)
```

```python
stop_words=set(stopwords.words('english'))

filtered=(word for word in tokens if word.lower() not in stop_words)

print("Filtered tokens:",list(filtered))
```

```python
nltk.download('averaged_perceptron_tagger_eng')
pos_tags=nltk.pos_tag(tokens)
print("pos tags(nltk):",pos_tags)
```

```python
import nltk
import spacy

nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer

text = "The boys are playing games and running faster than before."

tokens=word_tokenize(text.lower())
stemmer = PorterStemmer()

print("NLTK Stemming Output:")
for word in tokens:
    print(word, "->", stemmer.stem(word))


nlp=spacy.load("en_core_web_sm")
doc = nlp(text)

print("\nspaCy Morphological Analysis:")
for token in doc:
    print(token.text, "->", token.lemma_, "|", token.pos_, "|", token.morph)
```
