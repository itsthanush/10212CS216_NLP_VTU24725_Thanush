1.   the Transformers library
2.   stopwords as delimiters using NLTK and spaCy



### Part A: Tokenization using Transformers

```python
# 1. Import the tokenizer from the transformers library
from transformers import BertTokenizer

# 2. Load a pretrained tokenizer (e.g., BERT tokenizer)
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# 3. Input the sample text
sample_text = "Natural Language Processing is an exciting field of Artificial Intelligence."

# 4. Tokenize the text into subwords/tokens
tokens = tokenizer.tokenize(sample_text)

# 5. Display the tokens
print("Original Text:", sample_text)
print("Tokens:", tokens)
```

### Part B: Tokenization using Stopwords (NLTK)

```python
# 1. Import required libraries
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# 2. Download necessary datasets (run once)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab') # Added to resolve LookupError

# 3. Input the sample text (using the same text as Part A for consistency)
sample_text = "Natural Language Processing is an exciting field of Artificial Intelligence."

# 4. Tokenize text into words
word_tokens = word_tokenize(sample_text)

# 5. Get English stopwords
stop_words = set(stopwords.words('english'))

# Remove stopwords
filtered_tokens = [word for word in word_tokens if word.lower() not in stop_words]

# 6. Display filtered tokens
print("Original Text:", sample_text)
print("All Tokens:", word_tokens)
print("Filtered Tokens (without stopwords):", filtered_tokens)
```

### Part C: Tokenization using spaCy

```python
# 1. Install spaCy (if not already installed) and download the English model
!pip install spacy
!python -m spacy download en_core_web_sm

# 2. Import spaCy
import spacy

# 3. Load spaCy English model
nlp = spacy.load('en_core_web_sm')

# 4. Input the sample text (using the same text as Part A for consistency)
sample_text = "Natural Language Processing is an exciting field of Artificial Intelligence."

# 5. Process the text using spaCy NLP pipeline
doc = nlp(sample_text)

# 6. Remove stopwords and display tokens
filtered_tokens_spacy = [token.text for token in doc if not token.is_stop and not token.is_punct]

print("Original Text:", sample_text)
print("Tokens (with stopwords and punctuation):", [token.text for token in doc])
print("Filtered Tokens (without stopwords and punctuation):", filtered_tokens_spacy)
```

*   **spaCy Filtered Tokens:** `['Natural', 'Language', 'Processing', 'exciting', 'field', 'Artificial', 'Intelligence']`
