```python
#task6

def laplace_smoothing(counts, vocab_size):
    # Calculate the total count of all events
    total_count = sum(counts)
    # Apply the Laplace smoothing formula to each count
    smoothed_probs = [(count + 1) / (total_count + vocab_size) for count in counts]
    return smoothed_probs
# Example usage:
# Define the counts of events (e.g., word frequencies in a corpus)
counts = [3, 2, 0, 1] # counts for "cat", "dog", "fish", "bird"
vocab_size = 4 # size of the vocabulary (number of unique words)
# Calculate the smoothed probabilities
smoothed_probs = laplace_smoothing(counts, vocab_size)
# Print the smoothed probabilities
print("Smoothed probabilities:", smoothed_probs)
```

### Aim:

Both Laplace Smoothing (Add-1 Smoothing) and Add-k Smoothing are techniques used in natural language processing (NLP) and other probabilistic models to address the problem of **zero-frequency issues**. When calculating probabilities from observed data, if a particular event (e.g., a word or a sequence of words) has not been observed in the training corpus, its estimated probability would be zero. This is problematic because it implies that the event is impossible, which is often not true and can lead to issues in models like Naive Bayes classifiers or language models.

**The primary aim of smoothing techniques is to:**
1.  **Assign a non-zero probability to unseen events:** Prevent probabilities from being exactly zero for events not present in the training data.
2.  **Adjust observed probabilities:** Slightly reduce the probabilities of seen events to 'reserve' some probability mass for unseen events, ensuring that the sum of all probabilities remains 1.

### Main Content (Implementation Details):

#### Laplace Smoothing (Add-1 Smoothing)

Laplace smoothing is a simple form of Add-k smoothing where `k=1`. The formula for the smoothed probability of an event `i` is:

$P_{smoothed}(event_i) = \frac{count(event_i) + 1}{N + V}$

Where:
*   `count(event_i)` is the observed frequency of `event_i`.
*   `N` is the total count of all events.
*   `V` is the size of the vocabulary (total number of unique possible events).

**Code Breakdown (`laplace_smoothing` function):**
1.  `total_count = sum(counts)`: Calculates the sum of all observed frequencies.
2.  `smoothed_probs = [(count + 1) / (total_count + vocab_size) for count in counts]`: Applies the Laplace smoothing formula to each count in the input list, effectively adding 1 to each event's count and `vocab_size` to the total count.

#### Add-k Smoothing

Add-k smoothing generalizes Laplace smoothing by using an arbitrary positive value `k` (usually between 0 and 1) instead of just 1. The formula is:

$P_{smoothed}(event_i) = \frac{count(event_i) + k}{N + k * V}$

Where:
*   `k` is the smoothing parameter.
*   Other variables are the same as in Laplace smoothing.

**Code Breakdown (`add_k_smoothing` function):**
1.  `total_count = sum(counts)`: Calculates the sum of all observed frequencies.
2.  `smoothed_probs = [(count + k) / (total_count + k * vocab_size) for count in counts]`: Applies the Add-k smoothing formula using the provided `k` value. When `k=1`, this function behaves identically to Laplace smoothing.

### Results:

For both the `laplace_smoothing` and `add_k_smoothing` functions (when `k=1` for the latter), with the input `counts = [3, 2, 0, 1]` and `vocab_size = 4`:

**Output:** `Smoothed probabilities: [0.4, 0.3, 0.1, 0.2]`

**Explanation of Result:**
*   **Original Counts:** We had counts of 3, 2, 0, and 1 for four different events. The event with count 0 would have had a probability of 0 without smoothing.
*   **Total Count:** Sum of original counts = 3 + 2 + 0 + 1 = 6.
*   **Smoothed Denominator:** `total_count + vocab_size` = 6 + 4 = 10 (for Laplace).
*   **Smoothed Numerators:** Original counts + 1 = 4, 3, 1, 2.
*   **Smoothed Probabilities:**
    *   For event 1 (count 3): (3 + 1) / (6 + 4) = 4 / 10 = 0.4
    *   For event 2 (count 2): (2 + 1) / (6 + 4) = 3 / 10 = 0.3
    *   For event 3 (count 0): (0 + 1) / (6 + 4) = 1 / 10 = 0.1
    *   For event 4 (count 1): (1 + 1) / (6 + 4) = 2 / 10 = 0.2

The results show that even the event with an original count of 0 now has a non-zero probability (0.1). The probabilities sum up to 1 (0.4 + 0.3 + 0.1 + 0.2 = 1.0), and the distribution is slightly flattened compared to unsmoothed probabilities, allocating some probability mass to potentially unseen events.
